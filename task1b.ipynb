```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_dataset = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_56.csv')

dataset = raw_dataset.copy()
```

```python
drop = ['Country Code', 'SpecialNotes', 'Unnamed: 5']
dataset = dataset.drop(columns=drop)

dataset.dropna(inplace=True)
```

```python
income_plot = dataset['IncomeGroup'].value_counts()
print(income_plot)
```

```python
plt.pie(income_plot, labels=income_plot.index, autopct='%1.2f%%', startangle=140)
plt.show()
```

```python
sns.barplot(x=income_plot.index, y=income_plot.values)
plt.xticks(rotation=90)
plt.ylim([0,100])
plt.show()
```

```python
region_plot = dataset['Region'].value_counts()
print(region_plot)
```

```python
sns.barplot(x=region_plot.index, y=region_plot.values)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(region_plot, labels=region_plot.index, startangle=140, autopct='%1.1f%%')
plt.show()
```
